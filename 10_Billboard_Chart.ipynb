```python
import pandas as pd
import matplotlib.pyplot as plt

billboard = pd.read_csv('../data/billboard_charts.csv', parse_dates=['date'])
```

```python
top_10 = billboard[billboard['rank'] == 1]['artist'].value_counts().head(10)
```

```python
plt.style.use('ggplot')
```

```python
top_10.plot(
    kind = 'barh',
    title = 'Artist with most weeks at #1',
    figsize = (8,6),
    edgecolor  = 'black',
    linewidth = 3
)
plt.xlabel('Weeks at #1')
```

```python
artists = billboard[billboard['song'] == 'Please Mr. Postman']['artist'].value_counts()
```

```python
colors = ['#6c5ce7', '#00cec9', '#74b9ff']
artists.plot(
    figsize = (6,6),
    kind = 'pie',
    colors = colors,
    explode = (0,0,0.1),
    title = 'Please Mr. Postman Versions'
)
```

```python
billboard[billboard['song'] == 'Blinding Lights'].plot(x = 'date', y = 'rank', figsize = (10,6), linewidth = 2)
plt.gca().invert_yaxis()
```

```python
jl = billboard[billboard['artist'] == 'John Lennon']['song'].value_counts().head(8)
wings = billboard[billboard['artist'] == 'Wings']['song'].value_counts().head(8)
gh = billboard[billboard['artist'] == 'George Harrison']['song'].value_counts().head(8)
rs = billboard[billboard['artist'] == 'Ringo Starr']['song'].value_counts().head(8)

fig, axs = plt.subplots(2,2, figsize = (14,8))
fig.suptitle('Beatles solo songs (By weeks on chart)', fontsize = 20, color = 'purple')
jl.plot(kind = 'barh', ax = axs[0][0], color = 'red')
axs[0][0].set_title('John Lennon', color = 'red')
wings.plot(kind = 'barh', ax = axs[0][1], color = 'orange')
axs[0][1].set_title('Wings', color = 'orange')
gh.plot(kind = 'barh', ax = axs[1][0], color = 'blue')
axs[1][0].set_title('George Harrison', color = 'blue')
rs.plot(kind = 'barh', ax = axs[1][1] ,color = 'green')
axs[1][1].set_title('Ringo Starr', color = 'green')
plt.tight_layout() ## To get rid of overlapping
```
